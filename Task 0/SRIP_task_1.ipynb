**CDF Sampling for Frechet districution**

CDF Sampling formula for Frechet distribution: 

CDF(x)=exp(-(x/α)^(-β))

```python
import matplotlib.pyplot as plt
import numpy as np

def calc_frechet_pdf(x,a,s):
    y = (a/s) * ((x/s)**(-1-a)) * np.exp(- (x/s)**(-a) )
    return y

x = np.linspace(0.001,4,100)
plt.plot(x,calc_frechet_pdf(x,1,1))
plt.title("PDF")
plt.show()
```

**CDF**

```python
def calc_cdf_frechet(x,a,b):
  return np.exp(-(x/a)**(-b))

plt.plot(x,calc_cdf_frechet(x,1,1))
plt.title("CDF")
plt.show()
```

**Inverse CDF Sampling for frechet distribution**


```python
plt.plot(calc_cdf_frechet(x,1,1),x)
plt.title("Inverse CDF")
plt.show()
```

**Frechet Distribution:**

y=(a/s) * ((x/s)**(-1-a)) * e(- (x/s)**(-a) ) where e =2.712
